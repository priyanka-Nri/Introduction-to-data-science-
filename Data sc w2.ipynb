```python
def AnalyzeFrance():
  matchesPlayed = df['home_team'].value_counts()['France']
  print(f'total number of matches Played by France is:{matchesPlayed}')
```

```python
AnalyzeFrance()
```

```python
def AnalyzeTowns():
  TownData = df['home_team'].value_counts()
  for _ in TownData.index:
    print(f'total number of matches Played by {_} is:{TownData[_]}')
```

```python
AnalyzeTowns()
```

```python
TownName = []
matchesPlayed = []
TownData = df['home_team'].value_counts()
for _ in TownData.index:
  if TownData[_] > 30:
    TownName.append(_)
    matchesPlayed.append(TownData[_])
```

```python
plt.bar(matchesPlayed,TownName)
plt.ylabel('TownName')
plt.xlabel('matchesPlayed')
plt.title('MatchesPlayed vs TownName')
plt.show()
```

```python
def matchesPlayedByTown():
  TownName = []
  matchesPlayed = []
  TownData = df['home_team'].value_counts()
  for _ in TownData.index:
    if TownData[_] > 30:
      TownName.append(_)
      matchesPlayed.append(TownData[_])
      plt.bar(matchesPlayed,TownName)
      plt.ylabel('TownName')
      plt.xlabel('matchesPlayed')
      plt.title('MatchesPlayed vs TownName')
      plt.show()
```
